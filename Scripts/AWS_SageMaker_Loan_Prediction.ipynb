###  Exploratory Data Analysis

```python
# Copying the dataset into the the new final_loan_data for backup

final_loan_data = loan_data
```

```python
loan_data.groupby(loan_data["Self_Employed"]).count()
```

```python
# Replacing missing value of Self-Employed with more frequent category

final_loan_data["Self_Employed"].fillna('No', inplace = True)
final_loan_data["Self_Employed"].isnull().sum()
```

```python
# Replacing missing value with MEAN for continuous variable (numerical)


final_loan_data["LoanAmount"].fillna(final_loan_data["LoanAmount"].mean(), inplace = True)
final_loan_data["Loan_Amount_Term"].fillna(final_loan_data["Loan_Amount_Term"].mean(), inplace = True)
```

```python
# Replacing missing value with MODE for categorical variable

final_loan_data["Gender"].fillna(final_loan_data["Gender"].mode()[0], inplace = True)
final_loan_data["Married"].fillna(final_loan_data["Married"].mode()[0], inplace = True)
final_loan_data["Dependents"].fillna(final_loan_data["Dependents"].mode()[0], inplace = True)
final_loan_data["Credit_History"].fillna(final_loan_data["Credit_History"].mode()[0],inplace = True)
```

```python
# After replacing the missing value, rechecking the dataset to check if there is any missing value
final_loan_data.isnull().sum()
```

## Train Test Split Data

```python
train_data, test_data = np.split(final_loan_data.sample(frac=1, random_state=1729), [int(0.7 * len(final_loan_data))])
print(train_data.shape, test_data.shape)
```

```python
pd.concat([train_data['Loan_Status'], train_data.drop(['Loan_Status'], axis=1)], axis=1).to_csv('train.csv', index=False, header=False)
boto3.Session().resource('s3').Bucket(bucket_name).Object(os.path.join(prefix, 'train/train.csv')).upload_file('train.csv')
s3_input_train = sagemaker.inputs.TrainingInput(s3_data='s3://{}/{}/train'.format(bucket_name, prefix), content_type='csv')
```

```python
from sagemaker.amazon.amazon_estimator import get_image_uri 
from sagemaker.session import s3_input, Session
```

```python
pd.concat([test_data['Loan_Status'], test_data.drop(["Loan_Status"], axis=1)], axis=1).to_csv('test.csv', index=False, header=False)
boto3.Session().resource('s3').Bucket(bucket_name).Object(os.path.join(prefix, 'test/test.csv')).upload_file('test.csv')
s3_input_test = sagemaker.inputs.TrainingInput(s3_data='s3://{}/{}/test'.format(bucket_name, prefix), content_type='csv')
```

## Building Models Xgboot Inbuilt Algorithm

```python
#this line automatically looks for the XGBoost image URI and builds an XGBoost container.
container=sagemaker.image_uris.retrieve("xgboost", my_region, "1.2-1")
print(container)
```

```python
# initialize hyperparameters
hyperparameters = {
        "max_depth":"5",
        "eta":"0.2",
        "gamma":"4",
        "min_child_weight":"6",
        "subsample":"0.7",
        "objective":"binary:logistic",
        "num_round":50
        }
```

```python
# construct a SageMaker estimator that calls the xgboost-container
from sagemaker.debugger import Rule, rule_configs
from sagemaker.session import TrainingInput
estimator = sagemaker.estimator.Estimator(image_uri=container, 
                                          hyperparameters=hyperparameters,
                                          role=sagemaker.get_execution_role(),
                                          instance_count=1, 
                                          instance_type='ml.m5.2xlarge', 
                                          volume_size=5, # 5 GB 
                                          output_path=output_path,
                                          sagemaker_session=sagemaker.Session(),
                                          rules=[Rule.sagemaker(rule_configs.create_xgboost_report())])
```

```python
estimator.fit({'train': s3_input_train,'validation': s3_input_test})
```

### Deploy Machine Learning Model As Endpoints

```python
xgb_predictor = estimator.deploy(initial_instance_count=1,instance_type='ml.m4.xlarge')
```

```python
test_data_array = test_data.drop(['Loan_Status'], axis=1) #load the data into an array

test_data_array
```

```python
from sagemaker.predictor import csv_serializer

test_data.drop(['Loan_Status', 'Loan_ID'], axis=1).to_csv('test_model.csv', index=False, header=False)

with open("test_model.csv") as f:
    test_data_csv = f.read()
#load the data into an array

#print(test_data_csv)

xgb_predictor.serializer = sagemaker.serializers.CSVSerializer() # set the serializer type

#xgb_predictor.content_type = 'text/csv' # set the data type for an inference


predictions = xgb_predictor.predict(test_data_csv).decode('utf-8') # predict!

predictions_array = np.fromstring(predictions[1:], sep=',') # and turn the prediction into an array

print(predictions_array.shape)

```

```python
#print(predictions_array)
from sklearn.metrics import accuracy_score
predictions_round = [round(value) for value in predictions_array]


print(predictions_round)

# evaluate predictions
accuracy = accuracy_score(test_data["Loan_Status"], predictions_round)
print("Accuracy: %.2f%%" % (accuracy * 100.0))

cm = pd.crosstab(index=test_data['Loan_Status'], columns=np.round(predictions_array), rownames=['Observed'], colnames=['Predicted'])
print(cm)


tn = cm.iloc[0,0]; fn = cm.iloc[1,0]; tp = cm.iloc[1,1]; fp = cm.iloc[0,1]; p = (tp+tn)/(tp+tn+fp+fn)*100
print("\n{0:<20}{1:<4.1f}%\n".format("Overall Classification Rate: ", p))
print("{0:<15}{1:<15}{2:>8}".format("Predicted", "0: Yes", "1 : No"))
print("Observed")
print("{0:<15}{1:<2.0f}% ({2:<}){3:>6.0f}% ({4:<})".format("0: Yes", tn/(tn+fn)*100,tn, fp/(tp+fp)*100, fp))
print("{0:<16}{1:<1.0f}% ({2:<}){3:>7.0f}% ({4:<}) \n".format("1 : No", fn/(tn+fn)*100,fn, tp/(tp+fp)*100, tp))
```
